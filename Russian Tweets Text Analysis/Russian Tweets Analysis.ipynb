```python
more_stopwords = ['rt','amp','say', 'new', 'amb','get','arra','ara','one','get','like','lis','alis','dont','thats','im','really','make','know','people','the']

stoplist = stopwords.words('english') + more_stopwords

def cleantext(y):
    #Remove numbers, convert to string, lowercase and remove punctuation
    y['contentclean'] = y['contentclean'].str.replace('\d+', '')
    y['contentclean'] = y['contentclean'].astype(str)
    y['contentclean'] = y['contentclean'].apply(lambda x: " ".join(x.lower() for x in x.split()))
    y['contentclean'] = y['contentclean'].apply(lambda x: x.translate(str.maketrans("", "", string.punctuation)))
    words2 = ' '.join(y['contentclean'])
    stoplist = stopwords.words2('english') + more_stopwords
    words = y['contentclean'].apply(lambda x: " ".join(x for x in x.split() if x not in stoplist))
    return(words)
```

```python
import collections

# Stopwords
stopwords = set(line.strip() for line in cleanRTroll)
stopwords = stopwords.union(set(['the','amp','the','this','rt','just','you','it','we','im','httpstcokpxtohfw']))


def topwords(string, number):
    # Instantiate a dictionary, and for every word in the file, 
    # Add to the dictionary if it doesn't exist. If it does, increase the count.
    wordcount = {}
    # To eliminate duplicates, remember to split by punctuation, and use case demiliters.
    for word in string.lower().split():
        word = word.replace(".","")
        word = word.replace(",","")
        word = word.replace(":","")
        word = word.replace("\"","")
        word = word.replace("!","")
        word = word.replace("","")
        word = word.replace("","")
        word = word.replace("*","")
        if word not in stopwords:
            if word not in wordcount:
                wordcount[word] = 1
            else:
                wordcount[word] += 1
    
    # Print most common word
    print("Top {} most common words:\n".format(number))
    word_counter = collections.Counter(wordcount)
    for word, count in word_counter.most_common(number):
        print(word, ": ", count)
```

```python
import operator

def tophashtags(string):
    # Hashtags & mentions
    tag_dict = {}
    for i in string.index:
        tweet_text = string.ix[i]['content']
        tweet = tweet_text.lower()
        tweet_tokenized = tweet.split()
        
        for word in tweet_tokenized:
            # Hashtags - tokenize and build dict of tag counts
            if (word[0:1] == '#' and len(word) > 1):
                key = word.translate(str.maketrans("",""))
                if key in tag_dict:
                    tag_dict[key] += 1
                else:
                    tag_dict[key] = 1
    
    # The 10 most popular tags and counts
    top_tags = dict(sorted(tag_dict.items(), key=operator.itemgetter(1), reverse=True)[:10])
    top_tags_sorted = sorted(top_tags.items(), key=lambda x: x[1])[::-1]
    print ('Top 10 hashtags:')
    print ('----------------')
    for tag in top_tags_sorted:
        print (tag[0], '-', str(tag[1]))
```

```python
def TweetsbyDate(String, Title):
    RightPublishData = String[['publish_date']] #Isolate publish date
    #Create month, year, day column
    RightPublishData['month'] = pd.DatetimeIndex(RightPublishData['publish_date']).month
    RightPublishData['year'] = pd.DatetimeIndex(RightPublishData['publish_date']).year
    RightPublishData['day'] = pd.DatetimeIndex(RightPublishData['publish_date']).day
    
    #Group tweets by date
    ByDate = RightPublishData.groupby(["year", "month","day"]).size().reset_index(name="Tweets")
    #Combine columns back to date and put in index
    ByDate['Date'] = pd.to_datetime(ByDate.year.astype(str)+' '+ByDate.month.astype(str)+' '+ByDate.day.astype(str)) 
    ByDate = ByDate.set_index('Date')
    #Plot
    ByDate.reset_index().plot(x='Date', y='Tweets')
    plt.title(Title)
    return(ByDate)
```

```python

def TopTweetDays(String):
    RightPublishData = String[['publish_date']] #Isolate publish date
    #Create month, year, day column
    RightPublishData['month'] = pd.DatetimeIndex(RightPublishData.loc[:,'publish_date']).month
    RightPublishData['year'] = pd.DatetimeIndex(RightPublishData.loc[:,'publish_date']).year
    RightPublishData['day'] = pd.DatetimeIndex(RightPublishData.loc[:,'publish_date']).day
    
    #Group tweets by date
    ByDate = RightPublishData.groupby(["year", "month","day"]).size().reset_index(name="Tweets")
    #Combine columns back to date and put in index
    ByDate['Date'] = pd.to_datetime(ByDate.year.astype(str)+' '+ByDate.month.astype(str)+' '+ByDate.day.astype(str)) 
    ByDate = ByDate.set_index('Date')
    ByDate = ByDate.sort_values(by=['Tweets'], ascending = False)
    return(ByDate)
```
